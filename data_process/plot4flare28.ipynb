```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['savefig.dpi'] = 50  #视图窗口 每英寸点数，与figsize=(3,1)中的1对应
```

```python
## 多生成一些模板（图片性质的多样性）
def gen_img_rand_samples(template,name,path):
    width=len(template)
    fig,ax=plt.subplots(figsize=(3,1))
    #alpha 透明度 S数据点大小
    ax.scatter(np.arange(width),template,color='black',alpha=0.7,s=3)
    ax.invert_yaxis()
    ax.axis('off')
    ax.margins(0, 0)
    fig.savefig(path + name + '.png')
```

```python
# 密度阈值0.25
flare = np.loadtxt("../dataset/flare28_scale_dataset.csv" ,dtype=float,delimiter = ',')
path = '../dataset/test/flare28/'
for i in range(0,len(flare)):
    gen_img_rand_samples(flare[i][:-1].astype('float64'),str(i),path)
```

```python
flare = np.loadtxt("../dataset/flare28_scale_dataset_label2.csv" ,dtype=float,delimiter = ',')
path = '../dataset/test/flare28_2/'
for i in range(0,len(flare)):
    if flare[i,-1]== 0:
        gen_img_rand_samples(flare[i][:-1].astype('float64'),str(i),path)
```

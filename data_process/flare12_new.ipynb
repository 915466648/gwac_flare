```python
import os
import numpy as np
from sklearn import preprocessing
from matplotlib import pyplot as plt
from scipy.signal import savgol_filter
from scipy import interpolate
import math
```

```python
dir = '/home/wamdm/xinli/competition/flare_dataset_12'
dirlist = os.listdir(dir)  # 返回指定的文件夹包含的文件或文件夹的名字的列表。
if "abstract" in dirlist:
    dirlist.remove('abstract')

filterdata = []
for index,file in enumerate(dirlist):
    filepath = os.path.join(dir, file) 
    data = np.loadtxt(filepath ,dtype=float,delimiter = ' ')
    sgscale = savgol_filter(data[:,1], 5, 2, mode= 'nearest')
    filterdata.append(sgscale)
#     x = np.linspace(1, len(data),len(data))
#     plt.figure(file,figsize=(20,5))
#     plt.scatter(x,data[:,1],s=20)
#     plt.scatter(x,sgscale,s=20)
```

```python
# 提取出耀发形状部分
flaresub = []
for i,a in enumerate(filterdata[0:-1]):
    m = np.argmin(a)
    if i ==1 :
        x = [m-70,m+300]
    elif i ==2:
        x = [m-90,m+400]
    elif i ==3:
        x = [m-20,m+150]
    elif i ==4:
        x = [m-90,m+400]
    elif i ==5:
        x = [m-30,m+150]
    elif i ==6:
        x = [m-50,m+200]
    elif i ==8:
        x = [m-80,m+300]
    elif i ==9:
        x = [m-30,m+250]
    elif i ==10:
        x = [m-30,m+200]
    else:
        x = [m-30,m+200]
    b = savgol_filter(a[x[0]:x[1]], 41, 2, mode= 'nearest')
#     取反，使得尖向上，在经过归一化后，就与用模板生成的信号相同了，都是0-1的尖向上的信号
    flaresub.append(-b)
    print(len(b))
    plt.figure(i,figsize=(20,5))
    plt.title(i)
    plt.scatter(np.linspace(1,len(a),len(a)),a,s=20)
    plt.scatter(np.linspace(x[0],x[1],len(b)),b,s=20)
    
```

```python
# 插值至564，再抽样至188，从而提升上升沿的密度
def normalization(data):
    _range = np.max(data) - np.min(data)
    return (data - np.min(data)) / _range

sub = []
for i,a in enumerate(flaresub):
    f = interpolate.interp1d(np.linspace(0,len(a),len(a)),a,kind='cubic')
    xnew = np.linspace(0,len(a),564)
    a = normalization(f(xnew))[::3]
    sub.append(a)
    plt.figure(i,figsize=(3,1))
    plt.title(i)
    plt.scatter(np.linspace(0,len(a),len(a)), a,s=3)
#     plt.plot(np.linspace(0,len(a),len(a)), a, linewidth=1,label=i)
```

```python
f188 = []
for item in sub:
    f188.append(np.insert(item, 188, 0, axis=0))
    plt.plot(np.linspace(0,len(item),len(item)), item, linewidth=1,label=i)
```
